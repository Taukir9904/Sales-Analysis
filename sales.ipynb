# Overall Sales

```python
final=df.groupby(['Zone'])['Sales'].sum().reset_index()
final['Sales']=(final['Sales']/10000000).round(2)
final.rename(columns={'Sales':'Overall_Sales_CR'},inplace=True)
final
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x='Zone',y='Overall_Sales_CR',data=final)
```

# Won Sales

```python
df['Won/Lost/In Process'].value_counts()
```

```python
won=df[df['Won/Lost/In Process']=='Won']
```

```python
won['Won/Lost/In Process'].value_counts()
```

```python
won_f=won.groupby(['Zone'])['Sales'].sum().reset_index()
won_f['Sales']=(won_f['Sales']/10000000).round(2)
won_f.rename(columns={'Sales':'Final_Sales_CR'},inplace=True)
won_f
```

```python
Final=pd.merge(final,won_f,how='inner',on='Zone')
```

```python
Final['Loss']=Final['Overall_Sales_CR']-Final['Final_Sales_CR']
Final['Won %']=(Final['Final_Sales_CR']/Final['Overall_Sales_CR']*100).round(2)
Final['Loss %']=(Final['Loss']/Final['Overall_Sales_CR']*100).round(2)
```

```python
def sale_(zone):
    print('Please Select Zone from Below --->')
    print('\n')
    print(np.unique(df['Zone']))
    print('\n')
    a=df[df['Zone']==zone]
    b=(a['Sales'].sum()/10000000).round(2)
    print('\n')
    print('--->  Overall Sales for,',zone, 'is',b,'Crore')
    #print('For Won Sales')
    won=df[df['Won/Lost/In Process']=='Won']
    aa=won[won['Zone']==zone]
    bb=(aa['Sales'].sum()/10000000).round(2)
    print('\n')
    print('--->  Won Sales for,',zone, 'is',bb,'Crore')
    
```

```python
sale_('Chennai')
```
